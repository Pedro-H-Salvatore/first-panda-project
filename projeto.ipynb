```python
import pandas as pd 
import numpy as np 

df_vendas = pd.read_csv("vendas_tech.csv", low_memory= False) 
# encoding - caractere expecial 
# sep - define o separador ds infromações 
# low_memory = False -> carregar a tabela inteira 
display(df_vendas)
df_gerente = pd.read_excel("gerentes_lojas.xlsx")
display(df_gerente)
```

```python
# inspeção 
display(df_vendas)
display(df_vendas.head(15))
display(df_vendas.tail(15)) 
display(df_vendas.sample(15))
display(df_vendas.info())
display(df_vendas.shape)
display(df_vendas.describe())
display(list(df_vendas.columns))

# head -> mostra as 5 priemrias linhas 
# tail -> mostra as 5 ultimas linhas
# sample -> mostra 5 linhas aleatorias do seu banco de dados
# info -> mostra o resumo da sua base de dados, como ripo de dados, colunas e linhas
# shape -> mostra a quantidade de linhas e tabelas 
# describe -> mostra dados de estatisticas sobre as tabelas numéricas
# colums -> mostra as colunas em formato de lista
# () -> passa quantas linhas quer ver
```

```python
# tratamentos 
display(df_vendas[['Loja', 'Produto','Qtd']])
# colunas
# df_analise = df_vendas[['ID_Pedido', 'Data', 'Loja', 'Produto', 'Preco_Unitario', 'Qtd', 'Cliente']]
# display(df_vendas) -> jeito mais dificl -> Selecionar cada tabela e não adicionar a não desejada

df_analise = df_vendas.drop(columns=['Data_Base'])
display(df_analise)
# drop -> vai excluir a tabela desejada -> jeito facil 


# nulos 
# df_analise = df_analise.dropna()
# display(df_analise) -> vai apagar as linhas que tiverem qualquer valor nulo

df_analise['Loja'] = df_analise['Loja'].fillna("Online")

# fillna -> vai preencher as linhas vazias com o parametro passado pelo parenteses "Online"

display(df_analise)

# tipos de dados 

df_analise['Data'] = pd.to_datetime(df_analise['Data'], format="%Y-%m-%d")

# pd.to_datetime -> vai converter o tipo de dado, nesse exemplo de str -> datetime 
# format -> passa o estilo da data --- |%Y -> ano | %m -> mês |  %d -> dia|

# Padronização 
df_analise['Loja'] = df_analise['Loja'].str.strip() #Remove os espaços em branco
df_analise['Loja'] = df_analise['Loja'].str.title() #Deixa a primeira letra de cada palavra maiúscula



# duplicatas
df_analise = df_analise.drop_duplicates(subset=['ID_Pedido']) # drop_duplicates -> Tira valores duplicados | subset -> seleciona o parametro de excluzão, ou seja, "Se possuir Id's iguais, exclua"

# Exibição da base de dados formatadas 

display(df_analise)
display(df_analise.info())
display(df_analise.shape)
```

```python
#Criar novas colunas

#Faturamento
df_analise['Faturamento'] = df_analise['Qtd'] * df_analise['Preco_Unitario'] #Multiplicação entre as tabelas

#Forma de venda 
df_analise['Forma_De_Venda'] = np.where(df_analise['Loja']=='Online', 'Online', 'Presencial')
#np.where() -> como se fosse um if  ---->  (    valor,             se verdaderio,   se falso)

#Região
display(list(df_analise['Loja'].unique()))

dic_regioes = {
    'São Paulo': 'Sudeste',
 'Belo Horizonte': 'Sudeste',
 'Online': 'Online',
 'Rio De Janeiro': 'Sudeste',
 'Salvador': 'Nordeste',
 'Recife': 'Nordeste',
 'Curitiba': 'Sul',
 'Porto Alegre': 'Sul'
}

df_analise['Regiao'] = df_analise['Loja'].map(dic_regioes)

#map -> ele substitui o valor da coluna conforme o parametro passado, nesse caso, se o nome da loja for igual a São Paulo (key), tera seu valor como Sudeste (value)




display(df_analise)

display(df_analise.isna().sum())
```
